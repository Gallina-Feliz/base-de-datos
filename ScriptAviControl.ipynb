# [dbo].[DeleteGalpon]

```sql
/****** Object:  StoredProcedure [dbo].[DeleteGalpon]    Script Date: 5/09/2024 7:31:37 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[DeleteGalpon]
    @Numero_Galpon smallint,
    @Nuevo_Numero_Galpon smallint -- Nuevo galpón al que se trasladarán las gallinas
AS
BEGIN
    BEGIN TRANSACTION;
    BEGIN TRY
        -- Verificar si el nuevo galpón existe
        IF NOT EXISTS (SELECT 1 FROM Galpones WHERE Numero_Galpon = @Nuevo_Numero_Galpon)
        BEGIN
            THROW 50000, 'El nuevo galpón especificado no existe.', 1;
        END

        -- Verificar si hay gallinas registradas en el galpón que se desea eliminar
        IF EXISTS (SELECT 1 FROM Gallinas WHERE Numero_Galpon = @Numero_Galpon)
        BEGIN
            -- Contar el número de gallinas en el galpón a eliminar
            DECLARE @CantidadGallinas int;
            SELECT @CantidadGallinas = COUNT(*) FROM Gallinas WHERE Numero_Galpon = @Numero_Galpon;

            -- Actualizar el número de gallinas en el nuevo galpón
            UPDATE Galpones
            SET Numero_Gallinas = Numero_Gallinas + @CantidadGallinas
            WHERE Numero_Galpon = @Nuevo_Numero_Galpon;

            -- Actualizar el número de gallinas en el galpón a eliminar
            UPDATE Galpones
            SET Numero_Gallinas = 0
            WHERE Numero_Galpon = @Numero_Galpon;

            -- Actualizar el galpón de las gallinas registradas
            UPDATE Gallinas
            SET Numero_Galpon = @Nuevo_Numero_Galpon
            WHERE Numero_Galpon = @Numero_Galpon;
        END

        -- Eliminar el galpón
        DELETE FROM Galpones WHERE Numero_Galpon = @Numero_Galpon;

        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        ROLLBACK TRANSACTION;
        -- Manejo de errores
        THROW;
    END CATCH;
END;
GO
```

# [dbo].[EliminarGallina]

```sql
/****** Object:  StoredProcedure [dbo].[EliminarGallina]    Script Date: 5/09/2024 7:31:37 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[EliminarGallina]
    @Codigo_Gallinas smallint
AS
BEGIN
    BEGIN TRANSACTION;
    BEGIN TRY
        -- Obtener el galpón actual de la gallina
        DECLARE @Numero_Galpon smallint;
        SELECT @Numero_Galpon = Numero_Galpon FROM Gallinas WHERE Codigo_Gallinas = @Codigo_Gallinas;

        -- Eliminar la gallina
        DELETE FROM Gallinas WHERE Codigo_Gallinas = @Codigo_Gallinas;

        -- Actualizar el número de gallinas en el galpón
        UPDATE Galpones
        SET Numero_Gallinas = Numero_Gallinas - 1
        WHERE Numero_Galpon = @Numero_Galpon;

        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        ROLLBACK TRANSACTION;
        THROW;
    END CATCH;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[ListarGalpones]    Script Date: 5/09/2024 7:31:37 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[ListarGalpones]
AS
BEGIN
    -- Seleccionar información básica de los galpones
    SELECT 
        g.Numero_Galpon,
        g.Tipo_Galpon,
        g.Numero_Gallinas,
        -- Información de alimentación (si existe)
        ISNULL(a.CantidadAlimento, 0) AS CantidadAlimento,
        ISNULL(a.TipoAlimento, 'N/A') AS TipoAlimento,
        ISNULL(a.Fecha, '1900-01-01') AS FechaAlimentacion,
        -- Información de huevos (si existe)
        ISNULL(SUM(h.Cantidad), 0) AS CantidadHuevos,
        ISNULL(MAX(h.Fecha), '1900-01-01') AS FechaHuevos,
        ISNULL(t.Tipo_Huevo, 'N/A') AS TipoHuevo
    FROM 
        Galpones g
    LEFT JOIN 
        Alimentacion a ON g.Numero_Galpon = a.Numero_Galpon
    LEFT JOIN 
        Huevo h ON g.Numero_Galpon = h.Numero_Galpon
    LEFT JOIN 
        Tipo_Huevo t ON h.Id_TipoHuevo = t.Id_TipoHuevo
    GROUP BY 
        g.Numero_Galpon, 
        g.Tipo_Galpon, 
        g.Numero_Gallinas,
        a.CantidadAlimento,
        a.TipoAlimento,
        a.Fecha,
        t.Tipo_Huevo;
END;
GO
```

# [dbo].[ListarTipoHuevoConConteo]

```sql
/****** Object:  StoredProcedure [dbo].[ListarTipoHuevoConConteo]    Script Date: 5/09/2024 7:31:37 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[ListarTipoHuevoConConteo]
AS
BEGIN
    SELECT 
        TH.Id_TipoHuevo,
        TH.Tipo_Huevo,
        ISNULL(SUM(H.Cantidad), 0) AS Cantidad_Total_Huevos
    FROM Tipo_Huevo TH
    LEFT JOIN Huevo H ON TH.Id_TipoHuevo = H.Id_TipoHuevo
    GROUP BY TH.Id_TipoHuevo, TH.Tipo_Huevo;
END;
GO
```

# [dbo].[UpdateAlimentacion]

```sql
/****** Object:  StoredProcedure [dbo].[UpdateAlimentacion]    Script Date: 5/09/2024 7:31:37 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[UpdateAlimentacion]
    @Id_Alimentacion smallint,
    @CantidadAlimento smallint,
    @TipoAlimento varchar(100),
    @Fecha datetime,
    @Numero_Galpon smallint
AS
BEGIN
    -- Verificar que la alimentación existe
    IF NOT EXISTS (SELECT 1 FROM Alimentacion WHERE Id_Alimentacion = @Id_Alimentacion)
    BEGIN
        THROW 50000, 'El registro de alimentación especificado no existe.', 1;
    END

    -- Verificar que el galpón existe
    IF NOT EXISTS (SELECT 1 FROM Galpones WHERE Numero_Galpon = @Numero_Galpon)
    BEGIN
        THROW 50000, 'El galpón especificado no existe.', 1;
    END

    -- Actualizar la alimentación
    UPDATE Alimentacion
    SET 
        CantidadAlimento = @CantidadAlimento,
        TipoAlimento = @TipoAlimento,
        Fecha = @Fecha,
        Numero_Galpon = @Numero_Galpon
    WHERE 
        Id_Alimentacion = @Id_Alimentacion;
END;
GO
```

# [dbo].[UpdateGallina]

```sql
/****** Object:  StoredProcedure [dbo].[UpdateGallina]    Script Date: 5/09/2024 7:31:37 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[UpdateGallina]
    @Codigo_Gallinas smallint,
    @Id_Raza smallint,
    @Fecha_Nacimiento date,
    @Nuevo_Numero_Galpon smallint
AS
BEGIN
    BEGIN TRANSACTION;
    BEGIN TRY
        -- Obtener el galpón actual de la gallina
        DECLARE @Galpon_Anterior smallint;
        SELECT @Galpon_Anterior = Numero_Galpon 
        FROM Gallinas 
        WHERE Codigo_Gallinas = @Codigo_Gallinas;

        -- Verificar si se obtuvo el galpón anterior
        IF @Galpon_Anterior IS NULL
        BEGIN
            THROW 50001, 'La gallina con el código proporcionado no existe.', 1;
        END

        -- Actualizar la raza y el galpón de la gallina
        UPDATE Gallinas
        SET 
            Id_Raza = @Id_Raza,
            Fecha_Nacimiento = @Fecha_Nacimiento,
            Numero_Galpon = @Nuevo_Numero_Galpon
        WHERE 
            Codigo_Gallinas = @Codigo_Gallinas;

        -- Actualizar el número de gallinas en el galpón anterior si es diferente al nuevo
        IF @Galpon_Anterior <> @Nuevo_Numero_Galpon
        BEGIN
            UPDATE Galpones
            SET Numero_Gallinas = Numero_Gallinas - 1
            WHERE Numero_Galpon = @Galpon_Anterior;

            UPDATE Galpones
            SET Numero_Gallinas = Numero_Gallinas + 1
            WHERE Numero_Galpon = @Nuevo_Numero_Galpon;
        END

        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        ROLLBACK TRANSACTION;
        THROW;
    END CATCH;
END;
GO
```

# [dbo].[UpdateHuevo]

```sql
/****** Object:  StoredProcedure [dbo].[UpdateHuevo]    Script Date: 5/09/2024 7:31:37 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[UpdateHuevo]
    @Id_Huevo tinyint,
    @Cantidad smallint,
    @Id_TipoHuevo tinyint,
    @Numero_Galpon smallint
AS
BEGIN
    BEGIN TRANSACTION;
    BEGIN TRY
        -- Verificar que el galpón existe
        IF NOT EXISTS (SELECT 1 FROM Galpones WHERE Numero_Galpon = @Numero_Galpon)
        BEGIN
            THROW 50000, 'El galpón especificado no existe.', 1;
        END

        -- Verificar que el tipo de huevo existe
        IF NOT EXISTS (SELECT 1 FROM Tipo_Huevo WHERE Id_TipoHuevo = @Id_TipoHuevo)
        BEGIN
            THROW 50000, 'El tipo de huevo especificado no existe.', 1;
        END

        -- Actualizar el registro de huevo
        UPDATE Huevo
        SET Cantidad = @Cantidad,
            Id_TipoHuevo = @Id_TipoHuevo,
            Numero_Galpon = @Numero_Galpon
        WHERE Id_Huevo = @Id_Huevo;

        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        ROLLBACK TRANSACTION;
        -- Manejo de errores
        THROW;
    END CATCH;
END;
GO
```
